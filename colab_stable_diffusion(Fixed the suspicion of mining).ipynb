```python
#@title ## 2.1 Download List Management/CRUD Model or Other Files
#@markdown ####Selecting Mod Collection (Creating in Root Directory if Collection Not Found in Cloud Drive):##
json = "model.json" #@param {type: "string"}
# Create and Write Text to File
json_dir=''

#Recursive directory search for mod collection
def find_model_json(root_path,json):
    for file_name in os.listdir(root_path):
        file_path = os.path.join(root_path, file_name)
        if os.path.isdir(file_path):
            res=find_model_json(file_path,json)
            if res is not None:
              return res
        elif file_name == json:
            print("Found model.json file:", file_path)
            return file_path

if 'pan' in globals() and pan=="Allow":
  json_dir=find_model_json('/content/drive/',json)
  if json_dir==None:
    with open(f'/content/drive/MyDrive/{json}', 'w') as f:
      f.write('[]')
      json_dir=f'/content/drive/MyDrive/{json}'
else:
  json_dir=f'/content/{json}'

import json
import pandas as pd
# drive.mount('/content/drive')
from IPython.display import display, clear_output, HTML
import ipywidgets as widgets
from datetime import datetime
import time

custom_css = """
<style>
    .custom-button {
        padding: 10px 25px;
    }
    .custom-button.save {
        background-color: #67C23A;
        float:right
    }
    .custom-button.add {
        background-color: #409EFF;
    }
    .custom-button.delete {
        background-color: #F56C6C;
    }
    .custom-button.update {
        background-color: #E6A23C;
    }
</style>
"""
display(HTML(custom_css))

# Load JSON data as Python object
try:
    # Try opening your file.
    with open(json_dir, 'r') as f:
        try:
            # Try using the load() function from the json module to convert the file content to JSON.
            data = json.load(f)
        except json.JSONDecodeError:
            # If a JSONDecodeError exception occurs, set data as an empty dictionary.
            data = {}
except FileNotFoundError:
    # If a FileNotFoundError exception occurs, set data as an empty dictionary.
    data = {}

print(json_dir)
print(data)
# Convert JSON data to Pandas DataFrame
df = pd.DataFrame(data)

# Display table
def show_table():
    clear_output()
    # New Field Compatibility
    if not df.columns.empty and 'path' not in df.columns:
      df['path'] = ''
    styled_df = df.style.set_properties(**{'text-align': 'center'})

    # Set column width and center alignment
    styled_df.set_table_styles([{'selector': 'th', 'props': [('max-width', '150px'), ('text-align', 'center')]}])

    # import DataFrame
    display(styled_df)
    display(widgets.HBox([input_name, input_type, input_download_link, input_size]))
    display(widgets.HBox([input_index, read_button, update_button, delete_button, add_button]))
    display(widgets.HBox([save_button]))

# Add a row of data
def add_row(button):
  try:
    if not input_download_link.value:
      display(HTML(f"<p style='color: red;'>Download link is a required field</p>"))
      return
    new_row = {
        "name": input_name.value.strip() or 'unnamed',
        "type": 'undefined' if input_type.value == 'If not selected, it will be determined based on the downloaded content' else input_type.value,
        "downloadLink": input_download_link.value.strip(), # Modify the key name here
        "size": input_size.value,
        "path": input_path.value,
        "lastChangeDate": datetime.now().strftime('%Y-%m-%d')
    }
    global df
    df = df.append(new_row, ignore_index=True)
    lock_buttons()
    show_table()
    clear_input()
    display(HTML("<p id='add-msg' style='color: green;'>add{}successfully</p>".format(new_row)))
    time.sleep(1)
    display(HTML("<script>document.getElementById('add-msg').remove()</script>"))
  except Exception as e:
    display(HTML(f"<p style='color: red;'>addError：{e}</p>"))

# Delete a row of data
def delete_row(button):
  try:
    index = int(input_index.value)
    global df
    df = df.drop(index)
    lock_buttons()
    show_table()
    display(HTML("<p id='delete-msg' style='color: green;'>delete{}successfully</p>".format(index)))
    time.sleep(1)
    display(HTML("<script>document.getElementById('delete-msg').remove()</script>"))
  except Exception as e:
    display(HTML(f"<p style='color: red;'>deleteError：{e}</p>"))

#Read a row of data
def read_row(button):
    try:
        index = int(input_index.value)
        input_name.value = df.at[index, 'name']
        input_type.value ='If not selected, it will be determined based on the downloaded content' if df.at[index, 'type']=='undefined' else df.at[index, 'type']
        input_download_link.value = df.at[index, 'downloadLink']
        input_size.value = df.at[index, 'size']
        input_path.value = df.at[index, 'path']
        display(HTML(f"<p id='read-msg' style='color: green;'>read{index}successfully</p>"))
        time.sleep(1)
        display(HTML("<script>document.getElementById('read-msg').remove()</script>"))
    except Exception as e:
        display(HTML(f"<p style='color: red;'>readError：{e}</p>"))
# Modify a row of data
def update_row(button):
    try:
      index = int(input_index.value)
      if input_download_link.value:
        df.at[index, 'downloadLink'] = input_download_link.value.strip()
      else:
        display(HTML(f"<p style='color: red;'>Download link is a required field:{e}</p>"))
        return
      df.at[index, 'name'] = input_name.value.strip() or 'unnamed'
      df.at[index, 'type'] = 'undefined' if input_type.value == 'If not selected, it will be determined based on the downloaded content' else input_type.value
      df.at[index, 'size'] = input_size.value
      df.at[index, 'path'] = input_path.value or ''
      df.at[index, 'lastChangeDate'] = datetime.now().strftime('%Y-%m-%d')
      lock_buttons()
      clear_input()
      show_table()
      display(HTML("<p id='update-msg' style='color: green;'>edit successfully</p>".format(index)))
      time.sleep(1)
      display(HTML("<script>document.getElementById('update-msg').remove()</script>"))
    except Exception as e:
      display(HTML(f"<p style='color: red;'>editError：{e}</p>"))

#Clear input field
def clear_input():
    try:
        input_name.value = ''
        input_type.value = ''
        input_download_link.value = ''
        input_size.value = ''
        input_path.value = ''
    except Exception as e:
        display(HTML(f"<p style='color: red;'>Error resetting input box：{e}</p>"))

#save
def save_row(button):
    global df
    try:
        with open(json_dir, 'w') as Wfile:
            json=df.to_json(orient='records')
            Wfile.write(json)
            display(HTML("<p id='save-msg' style='color: green;'>save successfully</p>"))
            time.sleep(1)
            display(HTML("<script>document.getElementById('save-msg').remove()</script>"))
    except Exception as e:
        display(HTML(f"<p style='color: red;'>saveError：{e}</p>"))



# Define input field and button
input_name = widgets.Text(description='name',placeholder='Not filling it out will result in determination based on the downloaded content')
input_type = widgets.Text(description='type',placeholder='Optional')
input_download_link = widgets.Text(description='download link',placeholder='Required：civitai or huggingface') # Modify the description here
input_size = widgets.Text(description='size',placeholder='Optional')
input_path = widgets.Text(description='download path',placeholder='except normal model,others are required')
input_index = widgets.Text(description='mod index',placeholder='Used for CRUD operations, not for adding new content',layout=widgets.Layout(margin='10px'))
add_button = widgets.Button(description='add mod', layout=widgets.Layout(margin='10px 180px'), button_style='info')
delete_button = widgets.Button(description='delete mod', layout=widgets.Layout(margin='10px 10px'), button_style='danger')
update_button = widgets.Button(description='edit mod', layout=widgets.Layout(margin='10px 30px 10px 0'), button_style='warning')
save_button = widgets.Button(description='save', layout=widgets.Layout(margin='20px auto'), button_style='success')
read_button = widgets.Button(description='read mod', layout=widgets.Layout(margin='10px 0px 10px 30px'), button_style='primary')


#Throttling
def lock_buttons():
    # add_button.disabled = True
    delete_button.disabled = True
    update_button.disabled = True

def unlock_buttons():
    # add_button.disabled = False
    delete_button.disabled = False
    update_button.disabled = False

def on_input_change(change):
  if change.new:
    # len(change.new)
    if change.new==change.owner.value:
      if len(change.new)==1:
        lock_buttons()
        time.sleep(0.5)
        unlock_buttons()
  else:
    lock_buttons()


# Add an input event listener to the input field
# input_name.observe(on_input_change, names='value')
# input_type.observe(on_input_change, names='value')
# input_download_link.observe(on_input_change, names='value')
# input_size.observe(on_input_change, names='value')
input_index.observe(on_input_change, names='value')

add_button.on_click(add_row)
delete_button.on_click(delete_row)
update_button.on_click(update_row)
save_button.on_click(save_row)
read_button.on_click(read_row)

# Display a table and interactive buttons
lock_buttons()
show_table()
```
